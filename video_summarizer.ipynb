```python
# importing the youtube transcript api
from youtube_transcript_api import YouTubeTranscriptApi
import time
```

```python
# get video id
def get_video_id(url):
    if "v=" in url:
        return url.split("v=")[1].split("&")[0]
    elif "youtu.be/" in url:
        return url.split("youtu.be/")[1].split("?")[0]
    elif "youtube.com/shorts/" in url:
        return url.split("youtube.com/shorts/")[1].split("?")[0]
    else:
        raise ValueError("Invalid YouTube URL") 
```

```python
video_id = get_video_id("https://www.youtube.com/watch?v=Skj-d78j4YM&list=PLLalUvky4CLK3oT1DKNPagd_lTXooYVlR")
```

```python
def get_transcript(video_id):
    try:
        # List all available transcripts
        transcript_list = YouTubeTranscriptApi.list_transcripts(video_id)
        for transcript in transcript_list:
            print(f"Language: {transcript.language}, Auto-generated: {transcript.is_generated}")
            # Try to fetch the first available transcript
            return transcript.fetch()
        print("No transcripts found.")
        return []
    except Exception as e:
        print(f"Error: {e}")
        return []

transcript = get_transcript(video_id)
if transcript:
    print("First segment:", transcript[0])
else:
    print("Could not retrieve any transcript for this video.")
```

```python
#joining transcript text together

transcript_joined = " ".join([t['text'] for t in transcript])
prompt = f"Summarize this text: {transcript_joined}"
```
